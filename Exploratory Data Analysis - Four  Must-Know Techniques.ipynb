```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("humans.csv")
data
```

```python
# Box plot of the height column
plt.figure(figsize=(8, 6))
sns.boxplot(data['Height'], color='orange')
plt.title('Box plot of Height')
plt.xlabel('Height')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x= data['Height'], 
                y= data['Weight'], 
                data=data, color='blue')
plt.title('Scatter plot of Height vs. Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
# Pairplot of Height, Weight, and Age
sns.pairplot(data[['Height', 'Weight', 'Age']])
plt.show()
```

```python
# Creating a new variable BMI (Body Mass Index)
data['BMI'] = data['Weight'] / ((data['Height'] / 100) ** 2)
data.head()
```
